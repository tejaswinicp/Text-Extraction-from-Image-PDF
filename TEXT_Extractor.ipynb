# This notebook contains implementation of various techniques that an be used to extract text from PDF/Images.

Method#1: This method can be used for both Handwritten and Non Standard Form in the Sample test folder provided.

This involves converting a PDF file to image and further recognizing text from the images using Pytesseract. 

```python
from PIL import Image 
import pytesseract 
import sys 
from pdf2image import convert_from_path 
import os 
```

```python
# Path of the pdf
#For directories containing multiple PDF's, can iterate over all the PDF's using Glob
PDF_file = "Invoice.pdf"
  
''' 
Step #1 : Converting PDF to images 
'''
  
# Store all the pages of the PDF  
pages = convert_from_path(PDF_file, 500) 
  
# Counter to store images of each page of PDF to image 
image_counter = 1
  
# Iterate through all the pages stored above 
for page in pages: 
  
    # For each page, filename will be: 
    # PDF page n -> page_n.jpg 
    filename = "page_"+str(image_counter)+".jpg" 
    page.save(filename, 'JPEG') 
    image_counter = image_counter + 1
```

```python
''' 
Step #2 - Recognizing text from the images using OCR 
'''
    
# Total number of pages 
no_pages = image_counter-1
  
# Creating a text file to write the output 
outfile = "Extracted_Text.txt"
  
# Extracted text are added to the below file 
f = open(outfile, "a") 
  
# Iterate from 1 to total number of pages 
for i in range(1, no_pages + 1): 
  
    filename = "page_"+str(i)+".jpg"
          
    # Recognize the text as string in image using pytesseract 
    text = str(((pytesseract.image_to_string(Image.open(filename))))) 
    text = text.replace('-\n', '')     
    f.write(text) 
  
# Close the file after writing all the text. 
f.close() 
```

Method#2: Using PyPDF2 and textract

```python
import PyPDF2
import textract
```

```python
# This function will extract and return the pdf file text content.
def extractPdfText(filePath=''):

    fileObject = open(filePath, 'rb')
    pdfFileReader = PyPDF2.PdfFileReader(fileObject)
    totalPageNumber = pdfFileReader.numPages
    print('This pdf file contains totally ' + str(totalPageNumber) + ' pages.')

    currentPageNumber = 1
    text = ''

    # Loop in all the pdf pages.
    while(currentPageNumber < totalPageNumber ):

        # Get the specified pdf page object.
        pdfPage = pdfFileReader.getPage(currentPageNumber)

        # Get pdf page text.
        text = text + pdfPage.extractText()

        # Process next page.
        currentPageNumber += 1

    if(text == ''):
        # If can not extract text then use ocr lib to extract the scanned pdf file.
        text = textract.process(filePath, method='tesseract', encoding='utf-8')
      
    return text

extracted_text = extractPdfText("Invoice.pdf")
#Displaying only few lines of text
print(str(extracted_text[1:350]))
```

#Method#3: Optical Character Recognition wih Tesseract for extraction of text in images. Suitably for Standard form given in the Sample test folder

```python
import os
import tempfile
import subprocess

def ocr(path):
    temp = tempfile.NamedTemporaryFile(delete=False)

    process = subprocess.Popen(['tesseract', path, temp.name], stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
    process.communicate()

    with open(temp.name + '.txt', 'r') as handle:
        contents = handle.read()

    os.remove(temp.name + '.txt')
    os.remove(temp.name)

    return contents

str = ocr('/home/tejaswinicp/Desktop/Dental1.tif')
print(str)
```
